## MBS - Reg. Lin.

```python
def mbs(m, b, x, y, lr, n): 
  '''
  m = coef. ang.
  lr = learning rate
  n = num. de iteracoes
  '''
  for _ in range(n):
    mg = 0 # gradiente ang.
    bg = 0 # grandiente b
    for xi, yi in zip(x, y):
      xi, yi = float(xi), float(yi)
      mg += -(2/len(x)) * xi * (yi - (m * xi + b))
      bg += -(2/len(x)) * (yi - (m * xi + b))
    m = m - mg * lr
    b = b - bg * lr
  s = np.std(y - m * x + b)
  return m, b, s
```
